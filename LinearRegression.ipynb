```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with missing values
data = {'A': [1, 2, np.nan, 4, 5],
        'B': [10, 20, 30, np.nan, 50],
        'C': [100, 200, 300, 400, np.nan]}

df = pd.DataFrame(data)

# Function to fill missing values with random values within the range of available data
def fill_missing_with_random(column):
    not_null_values = column.dropna()
    null_count = column.isnull().sum()
    random_values = np.random.choice(not_null_values, null_count)
    column[column.isnull()] = random_values
    return column

# Apply the function to each column in the DataFrame
df_filled = df.apply(fill_missing_with_random)

# Display the original and filled DataFrame
print("Original DataFrame:")
print(df)
print("\nDataFrame with Missing Values Filled with Random Values:")
print(df_filled)
```
